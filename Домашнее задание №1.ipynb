С помощью регулярных выражение очистите текст от тэгов (мета информацию внутри тэгов тоже удалите). В получившемся тексте точки не должны слипаться с началом следующего предложений, также не должно быть лишних пробелов.

```python
def remove_tags(text):
    clear_text = re.sub('</.+?>', '', text) 
    one_space = re.sub('<.+?>', ' ', clear_text)
    return re.sub(' +', ' ', one_space)
clear_text = remove_tags(text)

print(clear_text)
```

## 4. Pymorphy vs Mystem (2 балла)

Лемматизируйте очищенный текст с помощью mystem и pymorphy. Сравните результаты: есть ли разница, что лучше. 

Для пайморфи используйте токенизацию из mystem, чтобы исключить влияние токенизации на результат.

```python
from pymystem3 import Mystem
from pymorphy2 import MorphAnalyzer
```

```python
mystem = Mystem()
```

```python
words_analized = mystem.analyze(clear_text)
mystem_lem = mystem.lemmatize(clear_text)
```

```python
tokens_mystem = [word['text'] for word in words_analized]
```

```python
morph = MorphAnalyzer()
```

```python
pymorphy_lem = [morph.parse(token['text'])[0].normal_form for token in mystem.analyze(clear_text)]
```

```python
from tabulate import tabulate

table = []

for i in range(0, len(mystem_lem)):
    if mystem_lem[i] != pymorphy_lem[i]:
        table.append([' '.join(tokens_mystem[i-4:i+4]).replace('\n', ' '), format(tokens_mystem[i]), mystem_lem[i], pymorphy_lem[i]])

print(tabulate(table, headers=['Контекст', 'Токен', 'Лемма Mystem', 'Лемма Pymorphy'], tablefmt='orgtbl', showindex="always"))
```

__Я считаю, что Mystem лучше подходит для лемматизации. При определении леммы учитывается контекст, что позволяет во многих случаях разрешать омонимию.
У pymorphy ощутимо больше ошибок (деньги - деньга). Тем не менее, есть случаи, когда mystem ошибался, а pymorphy - нет: при токене "ст__о__ит" mystem ошибочно выдал лемму "стоять".__

## 6*. Дизамбигуация в Mystem (2 балла)

*необязательное задание на 10 баллов

Одна из фичей майстема - контекстная дизамбигуация (т.е. определение правильного разбора в зависимости от контекста).

Запустите mystem на очищенном тексте с дизамбигуацией и без и проверьте есть ли какие-то различия.

```python
with_disamb = Mystem(disambiguation=True)
dis_lemmas = with_disamb.lemmatize(clear_text)
''.join(dis_lemmas)
```

```python
without_disamb = Mystem(disambiguation=False)
lemmas = without_disamb.lemmatize(clear_text)
''.join(lemmas)
```

```python
table = []

for i in range(0, len(dis_lemmas)):
    if dis_lemmas[i] != lemmas[i]:
        table.append([' '.join(tokens_mystem[i-4:i+4]).replace('\n', ' '), format(tokens_mystem[i]), dis_lemmas[i], lemmas[i]])

print(tabulate(table, headers=['Контекст', 'Токен', 'С дизамбигуацией', 'Без дизамбигуации'], tablefmt='orgtbl', showindex="always"))
```

__Несмотря на то, что при контекстной дизамбигуации все равно возникают ошибки, можно обратить внимание, что большинство форм стали правильными: говорящим - говорящий - говорить (при условии, что мы придерживаемся точки зрения, что причастие - это особая форма глагола, с начальной формой в И.п., ед.ч. м.р.) или Стоит - стоить - стоять. При это при дизамбигуации не меняется сравнительная форма, что странно: выше - выше - высоко.__
